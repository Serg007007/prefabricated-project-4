```python
def comparison(a1, a2, a3, b1):
    """—Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –æ—Ü–µ–Ω–∫–∏ —ç–∫—Å–ø–µ—Ä—Ç–æ–≤ –∏ –¥–æ–ª—é –ª—é–¥–µ–π, –ø–æ–¥—Ç–≤–µ—Ä–¥–∏–≤—à–∏—Ö, 
    —á—Ç–æ –æ–ø–∏—Å–∞–Ω–∏–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ä–µ–π—Ç–∏–Ω–≥ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è. 
    –ï—Å–ª–∏ –¥–∞–Ω–Ω—ã–µ –æ –¥–æ–ª–µ —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏—è –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç –≤ –¥–∞–Ω–Ω—ã—Ö —Ç–æ —Ä–µ–π—Ç–∏–Ω–≥ —Å—á–∏—Ç–∞–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –ø–æ –æ—Ü–µ–Ω–∫–∞–º —ç–∫—Å–ø–µ—Ä—Ç–æ–≤"""
    if math.isnan(b1):
        return round(0.25 * np.mean([a1,a2,a3]), 2)
    else: 
        return round(0.15 * np.mean([a1,a2,a3]) + 0.4 * b1, 2)    
```

### –õ–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏—è –æ–ø–∏—Å–∞–Ω–∏–π

```python
#spacy.cli.download("en_core_web_lg")
nlp = spacy.load("en_core_web_lg", disable=['parser', 'ner'])
```

```python
def lemmatizered(corpus):
    ''' —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–ø–æ–ª–Ω—è–µ—Ç —Ç–æ–∫–µ–Ω–∏–∑–∞—Ü–∏–∏—é –∏ –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏—é –º–∞—Å—Å–∏–≤–∞ —Ç–µ–∫—Å—Ç–æ–≤'''
    corpus_new = []
    for sentence in corpus:
        sentence = sentence.replace('  ', ' ')
        doc = nlp(sentence)      
        corpus_new.append(" ".join([token.lemma_ for token in doc]))
    return corpus_new
```

```python
%%time
pattern = r'[^a-zA-Z0-9]'
df['clear_text'] = df['query_text'].apply(lambda x: re.sub(pattern, " ", x))
df['clear_text'] = lemmatizered(df['clear_text'])
df.head()
```

```python
stop = [
    "boy",
    "girl",
    "child",
    "baby",
    "kid",
    "babe",
    "infant",
    "teenager",
    "teeny",
    "juvenile",
    "teener",
    "stripling",
]
```

```python
def stop_in_text(text: str) -> bool:
    """–ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –∏—â–µ–º —Å–ª–æ–≤–∞ –∏–∑ —Å—Ç–æ–ø-–ª–∏—Å—Ç–∞ child_stop –≤ –∑–∞–¥–∞–Ω–Ω–æ–º —Ç–µ–∫—Å—Ç–µ text. 
    –í–æ–∑–≤—Ä–∞—â–∞–µ–º True, –µ—Å–ª–∏ –Ω–∞—à–ª–æ—Å—å —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ —Å–ª–æ–≤–æ"""
    sign = False
    for w in stop:
        sign |= text.lower().find(w) > -1
    return sign
```

```python
df["clear_text_stop"] = df.apply(
    lambda x: stop_in_text(x["clear_text"]), axis=1
)
```

```python
norm = transforms.Normalize(
    mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    norm,
]) 
```

–î–∞–Ω–Ω—ã–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –ø—Ä–∏–º–µ–Ω–∏–º—ã –∫ –º–æ–¥–µ–ª–∏ ResNet. –î–ª—è –¥—Ä—É–≥–∏—Ö –º–æ–¥–µ–ª–µ–π —É—Ç–æ—á–Ω—è–π—Ç–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –≤ –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏.
–î–ª—è —á—Ç–µ–Ω–∏—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –ø–æ–Ω–∞–¥–æ–±–∏—Ç—Å—è –±–∏–±–ª–∏–æ—Ç–µ–∫–∞ PIL:

```python
@lru_cache(maxsize=128, typed=False)
def vec_im(image, path):
    ''' —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∏–º—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è –∏ –∞–¥—Ä–µ—Å –ø–∞–ø–∫–∏ —Å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º 
    –Ω–∞ –≤—ã—Ö–æ–¥–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—é –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è '''
    return resnet(
        preprocess(
            Image.open(join(path, image)).convert('RGB')).unsqueeze(0)).flatten()
```

```python
class Model_improve(nn.Module):
    def __init__(self, n_in_neurons, n_hidden_neurons_1, n_hidden_neurons_2, n_out_neurons, 
                 dropout_1, dropout_2, activation_1, activation_2):
        super(Model_improve, self).__init__()
        
        self.fc1 = nn.Linear(n_in_neurons, n_hidden_neurons_1)        
        self.bn1 = nn.BatchNorm1d(n_hidden_neurons_1)
        self.dp1 = nn.Dropout(p=dropout_1)
        self.act1 = activation_1
        
        self.fc2 = nn.Linear(n_hidden_neurons_1, n_hidden_neurons_2)
        self.bn2 = nn.BatchNorm1d(n_hidden_neurons_2) 
        self.dp2 = nn.Dropout(p=dropout_2)
        self.act2 = activation_2    
     
        self.fc3 = nn.Linear(n_hidden_neurons_2, n_out_neurons)
        self.act3 = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x)
        x = self.bn1(x)
        x = self.dp1(x)
        x = self.act1(x)

        x = self.fc2(x)
        x = self.bn2(x)
        x = self.dp2(x)
        x = self.act2(x)
       
        x = self.fc3(x)
        x = self.act3(x)
        
        return x
```

```python
net = Model_improve(n_in_neurons=vector_size, 
                    n_hidden_neurons_1=64, 
                    n_hidden_neurons_2=32, 
                    n_out_neurons=1, 
                    dropout_1=0.1, 
                    dropout_2=0.1, 
                    activation_1=nn.LeakyReLU(), 
                    activation_2=nn.LeakyReLU())
```

```python
class CustomEarlyStopping():
    """
    –†–∞–Ω–Ω—è—è –æ—Å—Ç–∞–Ω–æ–≤–∫–∞, —á—Ç–æ–±—ã –æ—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫—É, –∫–æ–≥–¥–∞ –ø–æ—Ç–µ—Ä—è –Ω–µ —É–ª—É—á—à–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π —ç–ø–æ—Ö–∏.
    """
    def __init__(self, patience=5, min_delta=0):
        """
        :param patience: —Å–∫–æ–ª—å–∫–æ —ç–ø–æ—Ö –∂–¥–∞—Ç—å, –ø—Ä–µ–∂–¥–µ —á–µ–º –æ—Å—Ç–∞–Ω–æ–≤–∏—Ç—å—Å—è, –∫–æ–≥–¥–∞ –ø–æ—Ç–µ—Ä—è –Ω–µ —É–ª—É—á—à–∞–µ—Ç—Å—è
        :param min_delta: –º–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É loss/–º–µ—Ç—Ä–∏–∫–∞–º–∏, –µ—Å–ª–∏ —É–ª—É—á—à–µ–Ω–∏–π –Ω–µ—Ç, —Ç–æ –æ—Å—Ç–∞–Ω–æ–≤–∏–º –æ–±—É—á–µ–Ω–∏–µ
        """
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False

    def __call__(self, val_loss):
        if self.best_loss == None:
            self.best_loss = val_loss
        elif self.best_loss - val_loss > self.min_delta:
            self.best_loss = val_loss
            # reset counter if validation loss improves
            self.counter = 0
        elif self.best_loss - val_loss < self.min_delta:
            self.counter += 1
            print(f"INFO: Early stopping counter {self.counter} of {self.patience}")
            if self.counter >= self.patience:
                print('INFO: Early stopping')
                self.early_stop = True
                
                

print(net)
def init_weights(m):
        if isinstance(m, nn.Linear):
            torch.nn.init.kaiming_normal_(m.weight)
            m.bias.data.fill_(0.01)
            
            
net.apply(init_weights)
print(net)


loss = nn.MSELoss()
optimizer = torch.optim.Adam(net.parameters(), lr=1e-3)

num_epochs = 1000 #—Å—Ç–∞–≤–∏–º –∫–∞–∫–æ–µ –Ω–∏–±—É–¥—å –±–æ–ª—å—à–æ–µ —á–∏—Å–ª–æ, –≤—Å–µ —Ä–∞–≤–Ω–æ –¥–æ –Ω–µ–≥–æ –Ω–µ –¥–æ–π–¥–µ–º –∏–∑ –∑–∞ EarlyStopping


early_stopping = CustomEarlyStopping(patience=3, min_delta=0.002) #–≠—Ç–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –º–æ–∂–Ω–æ –ø–æ–¥–±–∏—Ä–∞—Ç—å
```

```python
dataset_train = torch.utils.data.TensorDataset(
    torch.FloatTensor(X_train), 
    torch.FloatTensor(Y_train)
)
dataset_valid = torch.utils.data.TensorDataset(
    torch.FloatTensor(X_valid), 
    torch.FloatTensor(Y_valid)
)

train_dataloader = DataLoader(dataset_train, 
                              batch_size=32, 
                              shuffle=True,
                              num_workers=0
                             )
valid_dataloader = DataLoader(dataset_valid, 
                             batch_size=32, 
                             num_workers=0
                            )
```

```python
for epoch in range(num_epochs):
    net.train()
    for batch in train_dataloader:
        x_train, y_train = batch 
        optimizer.zero_grad()
        
        preds = net.forward(x_train).flatten()
               
        loss_value = torch.sqrt(loss(preds, y_train))
        loss_value.backward()
        optimizer.step()
        
    if epoch % 5 == 0:
        predicted_temp = [] 
        with torch.no_grad():
            net.eval()
            for batch in valid_dataloader:
                x_valid, y_valid = batch                
                
                valid_preds = net.forward(x_valid).flatten()
                predicted_temp.append(valid_preds)
                RMSE_loss = torch.sqrt(loss(valid_preds, y_valid))
                
        predicted_temp = torch.cat(predicted_temp).detach().numpy()
        RMSE = mean_squared_error(torch.FloatTensor(Y_valid), predicted_temp, squared=False)
        early_stopping(RMSE)
        print(f"epoch:{epoch}, RMSE test: {RMSE}")
        
        if early_stopping.early_stop:
            print('Early Stoppning!!!')
            break
```

```python
def create_nn_model():
    optimizer = Adam()
    model = Sequential()    
    model.add(Dense(units = 128, 
                    kernel_initializer = "uniform", 
                    activation = "relu", 
                    input_dim = vector_size,
                    activity_regularizer = regularizers.l2(0.001)))
    #model.add(BatchNormalization(axis=1))
    model.add(Dropout(rate = 0.1))
    model.add(Dense(units = 128, 
                    kernel_initializer = "uniform", 
                    activation = "relu"))
    #model.add(BatchNormalization(axis=1))
    model.add(Dropout(rate = 0.1))
    model.add(Dense(units = 1, 
                    kernel_initializer = "uniform", 
                    activation = "relu"))
    model.compile(optimizer=optimizer, 
                  loss=tf.keras.metrics.mean_squared_error, 
                  metrics=[tf.keras.metrics.RootMeanSquaredError(name='rmse')])

    return model
```

```python
def image_search(text, test_img_vec, model, test_images_name):
    '''–ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ, –¥–µ–ª–∞–µ—Ç –µ–≥–æ –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—é
    –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–∞—Ä—Ç–∏–Ω–∫—É —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –º–µ—Ç—Ä–∏–∫–∏'''
    for word in stop:
        if word in text:
            return 'This image is unavailable in your country in compliance with local laws'
        
    text_lem = []
    for x in range(len(queries)):
        text_lem.append(re.sub(pattern, " ", queries[x]))
    text_lem = np.array(text_lem, dtype=object)
    text_lem = lemmatizered(text_lem)
    text_test_vec = tfvect.transform(text_lem)
    text_test_vec = text_test_vec.toarray()
    
    for n in range(7):
        text_test_vec = np.concatenate((text_test_vec, text_test_vec))
    text_test_vec = text_test_vec[0:100]
    X = np.concatenate((test_img_vec, text_test_vec), axis=1)  
    preds_val = model.predict(X).flatten()
    idx = np.argsort(preds_val)[-3:]
    name = test_images_name.loc[idx, 'image']
    return name
```

```python
def image_search_pytorch(text, test_img_vec, model, test_images_name):
    '''–ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ, –¥–µ–ª–∞–µ—Ç –µ–≥–æ –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—é
    –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–∞—Ä—Ç–∏–Ω–∫—É —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –º–µ—Ç—Ä–∏–∫–∏'''
    for word in stop:
        if word in text:
            return 'This image is unavailable in your country in compliance with local laws'
        
    text_lem = []
    for x in range(len(queries)):
        text_lem.append(re.sub(pattern, " ", queries[x]))
    text_lem = np.array(text_lem, dtype=object)
    text_lem = lemmatizered(text_lem)
    text_test_vec = tfvect.transform(text_lem)
    text_test_vec = text_test_vec.toarray()
    
    for n in range(7):
        text_test_vec = np.concatenate((text_test_vec, text_test_vec))
    text_test_vec = text_test_vec[0:100]
    X = np.concatenate((test_img_vec, text_test_vec), axis=1)
    model.eval()
    with torch.no_grad():
        X = torch.FloatTensor(X)        
        preds_val = net(X).flatten(start_dim=-2)        
        idx = np.argsort(preds_val)[-3:]
        preds_val = pd.DataFrame(preds_val.numpy(), columns = ['ratio'])
        name = test_images_name.loc[idx, 'image']
        ratio = preds_val.loc[idx, 'ratio']
    
   
    return name, ratio, idx
```

–¢–µ—Å—Ç–∏—Ä—É–µ–º –º–æ–¥–µ–ª—å –Ω–µ–π—Ä–æ—Å–µ—Ç–∏ pytorch

```python
count = 1
print(queries[0])
name_image, ratio, idx = image_search_pytorch(queries, test_tf_img, net, test_images)
idx = idx.numpy()
fig = plt.figure(figsize=(12, 12))

for i in idx:
    img = Image.open(join(test_path, name_image[i]))
    fig.add_subplot(1, 3, count)
    count += 1
    plt.imshow(np.array(img))
    plt.axis("off")   
    plt.title(ratio[i])
```
